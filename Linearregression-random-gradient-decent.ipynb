```python
import numpy as np
from statistics import mean
import matplotlib.pyplot as plt
from matplotlib import style
from random import randint,random
style.use('ggplot')
```

```python
def coefficients_sgd(xs,ys, l_rate, n_epoch):
    m=0
    b=0
    errors=[]
    for epoch in range(n_epoch):
        sum_error = 0
        for i  in range(len(xs)):
            r_index = randint(0,len(xs)-1)
            yhat = xs[r_index]*m+b
            error = yhat - ys[r_index]
            sum_error+=error
            b = b - l_rate * error
            m = m - l_rate * error * xs[r_index]
        errors.append(sum_error/len(xs))
    return m , b, errors
```

```python
l_rate = 0.001
n_epoch = 50
Xa = 2 * np.random.rand(100, 1)
ya = 4 + 3 * Xa + np.random.randn(100, 1)
X =[float(Xa[i]) for i in range(100)]
y = [float(ya[i]) for i in range(100)]
m,b,errors = coefficients_sgd(X,y,l_rate,n_epoch)
plt.plot(errors)
plt.show()
```

```python

x_p = [max(X)*(x/len(X)) for x in range(len(X)+1)]
y_p = [x_p[x]*m+b for x in range(len(X)+1)]
plt.scatter(X,y,color="green")
plt.plot(x_p,y_p,color="red")
plt.show()
```
