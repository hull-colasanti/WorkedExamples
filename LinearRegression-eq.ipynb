```python
import numpy as np
from statistics import mean
import matplotlib.pyplot as plt
from matplotlib import style
from random import randint,random
style.use('ggplot')
Xt = [random()*100for _ in range(100)]
yt = [(Xt[i]+random()*40)for i in range(100)]
X = np.array(Xt)
y = np.array(yt)
plt.scatter(X,y,color="green")
plt.show()
```

```python
def best_fit_slope_and_intercept(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    b = mean(ys) - m*mean(xs)
    return m, b
```

```python
m,b = best_fit_slope_and_intercept(X,y)
print(m,b)
```

```python
regresion_line = [0]*len(X)
for x in range(len(X)):
    regresion_line[x]=x*m+b 
```

```python
plt.scatter(X,y,color="green")
plt.plot(regresion_line,color="red")
plt.show()
```

```python
def predict(x, m,b):
    yhat = x*m+b
    return yhat
 
# Estimate linear regression coefficients using stochastic gradient descent
def coefficients_sgd(xs,ys, l_rate, n_epoch):
    m=2
    b=0
    for epoch in range(n_epoch):
        sum_error = 0
        for i  in range(len(xs)):
            random_index = np.random.randint(len(xs))
            yhat = predict(xs[random_index],m,b)
            error = yhat - ys[random_index]
            sum_error += error**2
            b = b - l_rate * error
            m = m - l_rate * error * xs[i]
    return m , b
```

```python
l_rate = 0.01
n_epoch = 5000
m,b = coefficients_sgd(Xt,yt, l_rate, n_epoch)
print(m,b)
```
